```python
import matplotlib.pyplot as plt
import numpy as np

# Define fuzzy membership functions for cordialidade (placeholders)
def cordialidade_media(x):
    if 3 <= x <= 5:
        return (x - 3) / (5 - 3)
    elif 5 < x <= 7:
        return (7 - x) / (7 - 5)
    else:
        return 0.0

def cordialidade_alta(x):
    if 6 <= x <= 8:
        return (x - 6) / (8 - 6)
    elif 8 < x <= 10:
        return 1.0
    else:
        return 0.0

# Placeholder for avaliar_atendimento (fuzzy inference system)
def avaliar_atendimento(tempo, cordialidade):
    # This function would typically implement fuzzy rules and defuzzification.
    # For now, return a dummy quality value based on a simple combination.
    # Assuming 'qualidade_final' is a single output value scaled from 0-10.
    quality = (tempo / 60.0) * 0.3 + (cordialidade / 10.0) * 0.7
    return {'qualidade_final': quality * 10}

# Implemented function for plotar_funcoes_pertinencia
def plotar_funcoes_pertinencia():
    print("Plotando funções de pertinência para Cordialidade.")
    cord_range = np.arange(0, 11, 0.1) # Use a finer range for smoother plots

    plt.figure(figsize=(8, 5)) # Create a new figure for this plot
    plt.plot(cord_range, [cordialidade_media(c) for c in cord_range], 'y-', label='Média', linewidth=2)
    plt.plot(cord_range, [cordialidade_alta(c) for c in cord_range], 'g-', label='Alta', linewidth=2)
    plt.title('Funções de Pertinência para Cordialidade')
    plt.xlabel('Pontuação (0-10)')
    plt.ylabel('Grau de Pertinência')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

# Call the plotting function for Cordialidade
plotar_funcoes_pertinencia()

# Superfície de qualidade (now in a separate figure to avoid plot conflicts)
plt.figure(figsize=(8, 6)) # Create a new figure for the surface plot
T, C = np.meshgrid(np.linspace(0, 60, 30), np.linspace(0, 10, 30))
Z = np.zeros_like(T)

for i in range(T.shape[0]):
    for j in range(T.shape[1]):
        resultado = avaliar_atendimento(T[i,j], C[i,j])
        Z[i,j] = resultado['qualidade_final']

contour = plt.contourf(T, C, Z, levels=20, cmap='RdYlGn')
plt.colorbar(contour)
plt.title('Superfície de Qualidade')
plt.xlabel('Tempo (min)')
plt.ylabel('Cordialidade')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Configurar gráficos
plt.style.use('default')
plt.rcParams['figure.figsize'] = (10, 6)

# Célula 2: Definir funções de pertinência para Tempo de Espera
def tempo_rapido(x):
  """Função de pertinência para tempo rápido (0-15 min)"""
  if x <= 0:
      return 0.0
  elif 0 < x <= 10:
    return 1.0
  elif 10 < x <= 15:
    return (15 - x) / (15 - 10)
  else:
    return 0.0

# Add plotting for tempo_rapido here
tempo_range = np.arange(0, 20, 0.1) # A range appropriate for tempo_rapido

plt.figure(figsize=(8, 5))
plt.plot(tempo_range, [tempo_rapido(t) for t in tempo_range], 'b-', label='Rápido', linewidth=2)
plt.title('Função de Pertinência para Tempo de Espera (Rápido)')
plt.xlabel('Tempo (min)')
plt.ylabel('Grau de Pertinência')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
